## Different thicknesses

```python
#----------------------------
THICKNESS_1_NM = 10.0               # thickness in nanometers
THICKNESS_2_NM = 20.0               # thickness in nanometers
THICKNESS_3_NM = 32.0               # thickness in nanometers
THICKNESS_4_NM = 90.0               # thickness in nanometers
V_START = -1.0                      # gate Start voltage
V_STOP = 1.0                        # gate Stop voltage
V_POINTS = 21
V_CNP = 0.0                         # Charge Neutrality Point (V), set to 0 for simplicity
RELATIVE_EPSILON = 3.5              # Relative permittivity
#----------------------------


thickness_m_1 = THICKNESS_1_NM * 1e-9
thickness_m_2 = THICKNESS_2_NM * 1e-9
thickness_m_3 = THICKNESS_3_NM * 1e-9
thickness_m_4 = THICKNESS_4_NM * 1e-9



capacitance_1 = (RELATIVE_EPSILON * epsilon_0) / thickness_m_1
capacitance_2 = (RELATIVE_EPSILON * epsilon_0) / thickness_m_2
capacitance_3 = (RELATIVE_EPSILON * epsilon_0) / thickness_m_3
capacitance_4 = (RELATIVE_EPSILON * epsilon_0) / thickness_m_4


gate_voltages = np.linspace(V_START, V_STOP, V_POINTS)
# Effective voltage (accounts for any offset at the charge neutrality point)
v_effective = gate_voltages - V_CNP

carrier_density_1 = (capacitance_1 * v_effective / e) * 1e-4
carrier_density_2 = (capacitance_2 * v_effective / e) * 1e-4
carrier_density_3 = (capacitance_3 * v_effective / e) * 1e-4
carrier_density_4 = (capacitance_4 * v_effective / e) * 1e-4


displacement_field_V_1 = (capacitance_1/(epsilon_0 * 2)) * v_effective * 1e-9
displacement_field_V_2 = (capacitance_2/(epsilon_0 * 2)) * v_effective * 1e-9
displacement_field_V_3 = (capacitance_3/(epsilon_0 * 2)) * v_effective * 1e-9
displacement_field_V_4 = (capacitance_4/(epsilon_0 * 2)) * v_effective * 1e-9
```

## Different gates

```python
# --- 1. Set Your Parameters Here ---
TOP_THICKNESS_NM = 10.0             # Top hBN thickness in nanometers
BOTTOM_THICKNESS_NM = 40.0          # Bottom hBN thickness in nanometers
V_TOP_START = -1.0                  # Top gate Start voltage
V_TOP_STOP = 1.0                    # Top gate Stop voltage
V_BOTTOM_START = -2.0               # Bottom gate Start voltage
V_BOTTOM_STOP = 2.0                 # Bottom gate Stop voltage
V_POINTS = 21                       # Number of voltage points
RELATIVE_EPSILON_HBN = 3.5          # Relative permittivity of hBN
RELATIVE_EPSILON_AL2O3 = 9.0        # Relative permittivity of Al2O3
V_TOP_CNP = 0.0                     # Top Charge Neutrality Point (V), set to 0 for simplicity
V_BOTTOM_CNP = 0.0                  # Bottom Charge Neutrality Point (V), set to 0 for simplicity
# -------------------------------------
```

```python
# --- 2. Calculations ---

# Convert thickness to meters
top_thickness_m = TOP_THICKNESS_NM * 1e-9
bottom_thickness_m = BOTTOM_THICKNESS_NM * 1e-9

# Create the voltage array
# Using np.linspace is safer for floating-point steps
top_gate_voltages = np.linspace(V_TOP_START, V_TOP_STOP, V_POINTS)
bottom_gate_voltages = np.linspace(V_BOTTOM_START, V_BOTTOM_STOP, V_POINTS)

# Effective voltage (accounts for any offset at the charge neutrality point)
v_top_effective = top_gate_voltages - V_TOP_CNP
v_bottom_effective = bottom_gate_voltages - V_BOTTOM_CNP

# Capacitance per unit area (F/m^2)
capacitance_top = (RELATIVE_EPSILON_HBN * epsilon_0) / top_thickness_m
capacitance_bottom = (RELATIVE_EPSILON_HBN * epsilon_0) / bottom_thickness_m

# Calculate carrier density (n)
# n = C * V_eff / e  (gives units of 1/m^2)
# Multiply by 1e-4 to convert 1/m^2 to 1/cm^2
carrier_density_top_cm2 = (capacitance_top * v_top_effective / e) * 1e-4
carrier_density_bottom_cm2 = (capacitance_bottom * v_bottom_effective / e) * 1e-4

# Calculate electric field (E) in V/nm
# This is the 'displacement_field' you calculated.

# E = V / d.  For E in V/nm, we use V / d_nm.
#electric_field_V_nm_top = v_top_effective* RELATIVE_EPSILON_HBN / TOP_THICKNESS_NM
#electric_field_V_nm_bottom = v_bottom_effective / BOTTOM_THICKNESS_NM
displacement_field_V_top = (capacitance_top/epsilon_0) * v_top_effective * 1e-9
displacement_field_V_bottom = (capacitance_bottom/epsilon_0) * v_bottom_effective * 1e-9


# ----- different gate materials ------
capacitance_hBN = (RELATIVE_EPSILON_HBN * epsilon_0) / top_thickness_m
capacitance_Al2O3 = (RELATIVE_EPSILON_AL2O3 * epsilon_0) / top_thickness_m

carrier_density_hBN_cm2 = (capacitance_hBN * v_top_effective / e) * 1e-4
carrier_density_Al2O3_cm2 = (capacitance_Al2O3 * v_top_effective / e) * 1e-4

displacement_field_hBN = (capacitance_hBN/epsilon_0) * v_top_effective * 1e-9
displacement_field_Al2O3 = (capacitance_Al2O3/epsilon_0) * v_bottom_effective * 1e-9

```

```python
# --- carrier density vs gate voltage plot ---
fig1, ax1 = plt.subplots(figsize= (5,3))
color_n = 'tab:red'
color_b = 'tab:blue'
ax1.set_xlabel('Gate Voltage (V)', fontsize=12)
ax1.set_ylabel(r'n  $[\frac{1}{cm^2}]$', fontsize=14)
ax1.plot(top_gate_voltages,carrier_density_top_cm2, color=color_n, linewidth=2.5)
ax1.plot(bottom_gate_voltages,carrier_density_bottom_cm2, color=color_b, linewidth=2.5)
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.legend(['top gate','bottom gate'])
plt.show()
```
